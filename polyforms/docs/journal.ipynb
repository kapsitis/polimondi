```python
from polyforms.grammar import *
from polyforms.polyiamond import Polyiamond
from polyforms.draw_scene import *
from polyforms.n_gon import *

# NEW_n6_9,a (bd)^k ce (df)^k d (fa)^k f (ba)^k b,6,8,4

axiom = 'a[P1]ce[P2]d[P3]f[P4]b'
rule0 = {'P1':'[P1]bd', 'P2': '[P2]df', 'P3': '[P3]fa', 'P4': '[P4]ba'}
g = Grammar(axiom, [rule0])
pp = [g.derive([0]*i) for i in range(8)]

print([p.lower() for p in pp])
print([len(p) for p in pp])
poly = [Polyiamond(p) for p in pp]

import sympy as sp

LL = [p.get_area() for p in poly]

x = sp.symbols('x')

# Use the first 5 points (k = 0..4) to build a degree<=4 interpolation polynomial
points = [(k, LL[k]) for k in range(5)]
P = sp.expand(sp.interpolate(points, x))   # exact (rational) Lagrange interpolation

print("Interpolation polynomial through LL[0..4]:")
print("  P(x) =", P)

# (sanity check) confirm the defining conditions P(k) == LL[k] for k = 0..4
assert all(P.subs(x, k) == LL[k] for k in range(5))

# Test the extrapolated 6th term
for k in [5,6,7]:
    pred = P.subs(x, k)
    print(f"\nP({x})  = {pred}")
    print(f"LL[{x}] = {LL[k]}")
    
    if pred == LL[k]:
        print(f"\nMATCH: the degree-4 polynomial also reproduces LL[{k}].")
    else:
        print(f"\nNO MATCH")
```
