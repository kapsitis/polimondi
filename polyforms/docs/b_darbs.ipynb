```python
from polyforms.grammar import *
from polyforms.polyiamond import Polyiamond
from polyforms.draw_scene import *
from polyforms.n_gon import *

# Grammar G_2 from bakalaura darbs
axiom = 'a[P1]cacd[P2]e[P2]fdfdfa[P1]b'
rule0 = {'P1':'ca[P1]', 'P2':'fd[P2]'}
g = Grammar(axiom, [rule0])
pp = [g.derive([0]*i) for i in range(3)]

print([p.lower() for p in pp])

print([len(p) for p in pp])
poly = [Polyiamond(p) for p in pp]

offsets = [(0,0), (0,0), (0,0)]

scene = DrawScene(Align.BASELINE)
for idx,poly in enumerate(poly):
    scene.add_polyiamond(f'p{idx}', poly, (offsets[idx]))

scene.pack()
scene.set_size_in(6,7)
```

```python
from polyforms.grammar import *
from polyforms.polyiamond import Polyiamond
from polyforms.draw_scene import *
from polyforms.n_gon import *

# \mathbf{S} &\to& \mathbf{P}_1\,\mathtt{ace}\,\mathbf{P}_2\,\mathtt{dfb}\,\mathbf{P}_1 \\
# (\mathbf{P}_1,\; \mathbf{P}_2) &\to& (\mathbf{P}_1\,\mathtt{ab},\; \mathbf{P}_2\,\mathtt{dede}) \\
# (\mathbf{P}_1,\; \mathbf{P}_2) &\to& (\varepsilon,\; \varepsilon)\\

axiom = '[P1]ace[P2]dfb[P1]'
rule0 = {'P1':'[P1]ab', 'P2': '[P2]dede'}
g = Grammar(axiom, [rule0])
pp = [g.derive([0]*i) for i in range(3)]

print([p.lower() for p in pp])

print([len(p) for p in pp])
poly = [Polyiamond(p) for p in pp]

offsets = [(0,0), (-20,0), (-75,0)]

scene = DrawScene(Align.BASELINE)
for idx,poly in enumerate(poly):
    scene.add_polyiamond(f'p{idx}', poly, (offsets[idx]))

scene.pack()
scene.show_grid()
scene.set_size_in(12,5)
```
