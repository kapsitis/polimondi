```python
from polyforms.polyiamond import Polyiamond
from polyforms.draw_scene import *
from polyforms.n_gon import *

magic_poly = Polyiamond([(12,'A'), (8,'B'), (2,'C'), (5,'B'), (11,'C'), (9,'D'), 
                         (3,'E'), (4,'D'), (10,'E'), (7,'F'), (1,'A'), (6,'F')])

mylist12 = get_all_solutions(list(range(12,0,-1)), Format.COMPACT)
mylist12_poly = [Polyiamond(a) for a in mylist12]
mylist12_areas = [p.get_area() for p in mylist12_poly]
max12 = max(mylist12_areas)

max_polygons_perfect = list(filter(lambda p: p.get_area() >= max12, mylist12_poly))

print('max_area is {}'.format(max12))
max_polygons = [magic_poly] + max_polygons_perfect

scene = DrawScene(Align.BASELINE)
for idx,poly in enumerate(max_polygons):
    scene.add_polyiamond(f'p{idx}', poly)

scene.pack()
scene.show_grid()
scene.set_size_in(11,5)
```
